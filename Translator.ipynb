```python
import pandas as pd
# Load data
file_path = 'parallel-corpus.xlsx'
df = pd.read_excel(file_path)

# Display the first 10 lines of Datset
df.head(10)
```

```python
# Remove extra columns, keeping only 'SENTENCES ' and 'MEANING'
df_cleaned = df[['SENTENCES ', 'MEANING']]

df_cleaned.head(10)
```

```python
import re
def remove_emoji_and_links(text):
    if not isinstance(text, str): # handle non-string values
        text = str(text)
    
    # Remove emojis using regex
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)

    # Remove links using regex
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    return text

# Apply the function to both English and Urdu columns
df_cleaned['Emoji_Removed_english'] = df_cleaned['SENTENCES '].apply(remove_emoji_and_links)
df_cleaned['Emoji_Removed_urdu'] = df_cleaned['MEANING'].apply(remove_emoji_and_links)

df_cleaned[['Emoji_Removed_english', 'Emoji_Removed_urdu']].head(10)
```

```python
import spacy
from collections import defaultdict

# Load spaCy English model
nlp_en = spacy.load("en_core_web_sm")

# Load the multilingual model
nlp = spacy.load('xx_ent_wiki_sm')
# Tokenization function using SpaCy
def tokenize_urdu(text):
    doc = nlp(text)
    return [token.text for token in doc]

# English tokenizer using spaCy
def tokenize_text_spacy(text):
    return [token.text for token in nlp_en(text)]

# Apply tokenization to English and Urdu column
df_cleaned['tokenized_english'] = df_cleaned['Emoji_Removed_english'].apply(tokenize_text_spacy)
df_cleaned['tokenized_urdu'] = df_cleaned['Emoji_Removed_urdu'].apply(tokenize_urdu)

# Function to create a word-to-index dictionary
def create_vocab(tokenized_sentences):
    word_to_index = defaultdict(lambda: len(word_to_index))
    word_to_index['<PAD>'] = 0  # Adding padding token
    for sentence in tokenized_sentences:
        for word in sentence:
            _ = word_to_index[word]  # Assign index
    return dict(word_to_index)

# Create vocabulary for English and URdu
english_vocab = create_vocab(df_cleaned['tokenized_english'])
urdu_vocab=create_vocab(df_cleaned['tokenized_urdu'])

# Display tokenized sentences and vocab size
print(df_cleaned[['tokenized_english','tokenized_urdu']].head(10))
print(f"English Vocab Size: {len(english_vocab)}")
print(f"Urdu Vocab Size: {len(urdu_vocab)}")
```

```python
import torch
import numpy as np

# Ensure df_cleaned has the required columns
required_columns = ['tokenized_english', 'tokenized_urdu']
if not all(col in df_cleaned.columns for col in required_columns):
    raise ValueError("DataFrame is missing required columns")

# Convert tokenized sentences to index sequences
def convert_to_indices(vocab, sequences):
    return [[vocab.get(word, 0) for word in sentence] for sentence in sequences]

english_sequences = convert_to_indices(english_vocab, df_cleaned['tokenized_english'])
urdu_sequences = convert_to_indices(urdu_vocab, df_cleaned['tokenized_urdu'])

# Convert sequences to PyTorch tensors
english_sequences = [torch.tensor(seq, dtype=torch.long) for seq in english_sequences]
urdu_sequences = [torch.tensor(seq, dtype=torch.long) for seq in urdu_sequences]

# Pad sequences using PyTorch's pad_sequence function
def pad_sequences_pytorch(sequences, maxlen, padding='post'):
    # First truncate sequences longer than maxlen
    truncated_sequences = [seq[:maxlen] if len(seq) > maxlen else seq for seq in sequences]
    
    # Use pad_sequence to pad them to the same length
    if padding == 'post':
        padded_sequences = torch.nn.utils.rnn.pad_sequence(truncated_sequences, batch_first=True, padding_value=0)
    else:  # If pre-padding is required
        padded_sequences = torch.nn.utils.rnn.pad_sequence(truncated_sequences, batch_first=True, padding_value=0)
        padded_sequences = torch.flip(padded_sequences, dims=[1])
    
    # Make sure the output shape is (num_sequences, maxlen)
    return padded_sequences[:, :maxlen]

# Set maximum sequence length
max_length = 20  

# Pad English and Urdu sequences
padded_english = pad_sequences_pytorch(english_sequences, maxlen=max_length, padding='post')
padded_urdu = pad_sequences_pytorch(urdu_sequences, maxlen=max_length, padding='post')

# Handle zero-length sequences directly by masking any rows that are all zeros
empty_mask = torch.all(padded_english == 0, dim=1)

# Apply mask to both padded_english and padded_urdu to ensure consistency
padded_english[empty_mask] = 0
padded_urdu[empty_mask] = 0

print(f"Shape of padded_english: {padded_english.shape}")
print(f"Shape of padded_urdu: {padded_urdu.shape}")

```

```python
import pandas as pd
import numpy as np


# Shuffle the dataset using random permutation
shuffled_indices = np.random.permutation(len(df))

# Calculate split indices
train_end = int(len(df_cleaned) * 0.8)  # 80% for training
val_end = train_end + int(len(df) * 0.1)  # 10% for validation

# Create training, validation, and test sets using the shuffled indices
train_indices = shuffled_indices[:train_end]
val_indices = shuffled_indices[train_end:val_end]
test_indices = shuffled_indices[val_end:]

train_df = df.iloc[train_indices]
val_df = df.iloc[val_indices]
test_df = df.iloc[test_indices]

print(f"Training set size: {len(train_df)}")
print(f"Validation set size: {len(val_df)}")
print(f"Test set size: {len(test_df)}")

# Reset indices if needed
train_df.reset_index(drop=True, inplace=True)
val_df.reset_index(drop=True, inplace=True)
test_df.reset_index(drop=True, inplace=True)
```

```python
x_train = padded_english[train_indices]  # Padded tokenized English training data
y_train = padded_urdu[train_indices]     # Padded tokenized Urdu training data

x_val = padded_english[val_indices]  # Validation English data
y_val = padded_urdu[val_indices]     # Validation Urdu data

x_test=padded_english[test_indices]
y_test=padded_urdu[test_indices]
# Get vocabulary sizes
english_vocab_size = len(english_vocab) + 1  # +1 for padding token
urdu_vocab_size = len(urdu_vocab) + 1        # +1 for padding token
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

# Parameters
embedding_dim = 40
rnn_units = 64

# Define the model
model = Sequential([
    Embedding(input_dim=english_vocab_size, output_dim=embedding_dim),
    SimpleRNN(units=rnn_units, return_sequences=True),
    Dense(units=urdu_vocab_size, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Display model summary
model.summary()

# Train the model
history = model.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=50, batch_size=10)

# Save the model
model.save('RNN_model.keras')

# Display accuracy results with % sign
train_accuracy = history.history['accuracy'][-1] * 100
val_accuracy = history.history['val_accuracy'][-1] * 100

print(f'Training Accuracy: {train_accuracy:.2f}%')
print(f'Validation Accuracy: {val_accuracy:.2f}%')
```

```python
import numpy as np
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
import gc  # For garbage collection

# Function to decode sequences back to words using the vocabulary
def decode_sequence(sequence, vocab):
    reverse_vocab = {index: word for word, index in vocab.items()}  # Reverse mapping
    return [reverse_vocab.get(index, '') for index in sequence if index != 0]

# Initialize list to store BLEU scores
bleu_scores = []
smoothing_function = SmoothingFunction()  # Create smoothing function instance

# Loop through each test sample
for i in range(len(x_test)):
    # Predict for one sample
    y_pred = model.predict(x_test[i:i+1])
    y_pred_sequence = np.argmax(y_pred, axis=-1)[0]  # Get predicted tokens

    # Decode the reference (actual) and predicted sequences
    reference = [decode_sequence(y_test[i], urdu_vocab)]
    candidate = decode_sequence(y_pred_sequence, urdu_vocab)
    
    # Calculate BLEU score with smoothing
    bleu_score = sentence_bleu(reference, candidate, smoothing_function=smoothing_function.method1)
    bleu_scores.append(bleu_score)
    
    # Clear memory
    gc.collect()

# Calculate average BLEU score
average_bleu = np.mean(bleu_scores)
print(f'Average BLEU Score: {average_bleu}')
```
